```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
time_data = {'Facebook':150, 'Instagram':100, 'Twitter':50, 'Snapchat':80}
social_media = list(time_data.keys())
time_spend = list(time_data.values())
```

```python
plt.figure(figsize=(5,5))
plt.bar(social_media, time_spend,width=0.5)
plt.xlabel('Social Media')
plt.ylabel('Time in minutes')
plt.title('Average time spend on social media')
```

```python
fb = {120, 90, 110, 100, 95}
ig = {80,90,75,85, 95}
tt = {50, 40, 55, 35, 45}
sc = {20, 30, 35, 25, 28}
```

```python
xpos = np.arange(len(fb))
xpos
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))
bandwidth = 0.2
plt.bar(xpos, fb, color='green', width= 0.3, label ='FB')
```

```python
style.use('ggplot')
plt.figure(figsize=(10,7))
bandwidth = 0.2
plt.bar(xpos, fb, color='green', width= 0.3, label ='FB')
plt.bar(xpos, ig, color='yellow', width= 0.3, label ='IG')
plt.bar(xpos, tt, color='red', width= 0.3, label ='TT')
plt.bar(xpos, sc, color='purple', width= 0.3, label ='SC')
```

```python
style.use('ggplot')
plt.figure(figsize=(12,7))
bandwidth = 0.2
plt.bar(xpos, fb, color='royalblue', width= 0.3, label ='FB')
plt.bar(xpos+0.2, ig, color='red', width= 0.3, label ='IG')
plt.bar(xpos+0.4, tt, color='green', width= 0.3, label ='TT')
plt.bar(xpos+0.6, sc, color='yellow', width= 0.3, label ='SC')
```
